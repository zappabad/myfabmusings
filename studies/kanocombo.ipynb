```python
def calculate_damage(chain, N, K, C):
    """
    Calculate the total damage dealt by a chain of spells.
    
    Parameters:
    chain (list of str): The sequence of cards played.
    N (int): The additional damage provided by Wildfire.
    
    Returns:
    int: The total damage dealt.
    """
    total_damage = 0
    cumulative_damage = 0
    spell_damage = 0
    resource_cost = 9
    
    for card in chain:
        if card == "dumb_dmg":
            resource_cost += 6
            spell_damage = C + cumulative_damage
            total_damage += spell_damage
            print(f"Dumb Damage cast: {spell_damage} damage dealt ({cumulative_damage} from previous wildfire pump), resource cost: {resource_cost}, total damage so far: {total_damage}")
        elif card == "first_wildfire":
            resource_cost += 7
            spell_damage = K + cumulative_damage
            total_damage += spell_damage
            
            print(f"First Wildfire cast: {spell_damage} damage dealt ({cumulative_damage} from previous wildfire pump), resource cost: {resource_cost}, total damage so far: {total_damage}")
            cumulative_damage += spell_damage
        elif card == "wildfire":
            resource_cost += 5
            spell_damage = N + cumulative_damage
            total_damage += spell_damage
            
            print(f"Wildfire cast: {spell_damage} damage dealt ({cumulative_damage} from previous wildfire pump), resource cost: {resource_cost}, total damage so far: {total_damage}")
            cumulative_damage += spell_damage
        elif card == "blazing":
            resource_cost += 3
            spell_damage = (total_damage) + cumulative_damage
            total_damage += spell_damage
            print(f"Blazing cast: {spell_damage} damage dealt ({cumulative_damage} from previous wildfire pump), resource cost: {resource_cost}, total damage so far: {total_damage}")
        elif card == "blazing_from_strategic":
            resource_cost += 7
            spell_damage = (total_damage) + cumulative_damage
            total_damage += spell_damage
            print(f"Blazing from Strategic cast: {spell_damage} damage dealt ({cumulative_damage} from previous wildfire pump), resource cost: {resource_cost}, total damage so far: {total_damage}")
    
    return total_damage

# Define the parameters
N = 4  # Example value for the damage boost from Wildfire
K = 4+3+2 # first wildfire damage, get all pumps in
C = 52

# Define the chains
chain1 = ["first_wildfire", "wildfire", "wildfire",
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          # extra blazings with strategic planning, adds 4 to the cost of each
          "blazing_from_strategic", "blazing_from_strategic", "blazing_from_strategic",
          "blazing_from_strategic", "blazing_from_strategic", "blazing_from_strategic",
          ]

# Define the chains
chain2 = ["dumb_dmg", #"dumb_dmg", "dumb_dmg",
          # blazings with remembrance
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          # extra blazings with strategic planning, adds 4 to the cost of each
          "blazing_from_strategic", "blazing_from_strategic", "blazing_from_strategic",
          "blazing_from_strategic", "blazing_from_strategic", "blazing_from_strategic",
          "blazing_from_strategic",
          ]

# Define the chains
chain3 = ["dumb_dmg", "dumb_dmg",# "dumb_dmg",
          # blazings with remembrance
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          # extra blazings with strategic planning, adds 4 to the cost of each
          "blazing_from_strategic", "blazing_from_strategic", "blazing_from_strategic",
          "blazing_from_strategic", "blazing_from_strategic", "blazing_from_strategic",
          "blazing_from_strategic",
          ]

# Define the chains
chain4 = ["dumb_dmg", "dumb_dmg", "dumb_dmg",
          # blazings with remembrance
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          "blazing", "blazing", "blazing",
          # extra blazings with strategic planning, adds 4 to the cost of each
          "blazing_from_strategic", "blazing_from_strategic", "blazing_from_strategic",
          "blazing_from_strategic", "blazing_from_strategic", "blazing_from_strategic",
          #"blazing_from_strategic",
          ]


# Calculate the total damage for each chain
#print(f"Chain 1: {chain1}")
#total_damage_chain1 = calculate_damage(chain1, N)
print(f"\nChain 2: {chain2}")
total_damage_chain2 = calculate_damage(chain2, N, K, C)
print("\n\n")
print(f"\nChain 3: {chain3}")
total_damage_chain3 = calculate_damage(chain3, N, K, C)
print("\n\n")
print(f"\nChain 4: {chain4}")
total_damage_chain4 = calculate_damage(chain4, N, K, C)
print("\n\n")
#total_damage_chain1,
total_damage_chain2, total_damage_chain3, total_damage_chain4
```
