```python
import requests
import pandas as pd

url = "https://the-fab-cube.github.io/flesh-and-blood-cards/json/english/card-flattened.json"

response = requests.get(url)
all_cards_df = pd.DataFrame(response.json())
```

```python
all_cards_df
```

```python
all_cards_df.columns
```

```python
set_id = 'UPR'
set_df = all_cards_df[all_cards_df['set_id'] == set_id]
unique_df = set_df.groupby(['name', 'pitch']).first().reset_index()
filtered_df = unique_df[~unique_df['types'].apply(lambda x: 'Ally' in x)]
filtered_df = filtered_df[~filtered_df['rarity'].apply(lambda x: 'F' in x)]
filtered_df = filtered_df[~filtered_df['rarity'].apply(lambda x: 'L' in x)]
filtered_df = filtered_df[~filtered_df['rarity'].apply(lambda x: 'T' in x)]
filtered_df
```
