```python
import pandas as pd

def run_clash_sim(deck_A, deck_B):

    # Create a DataFrame to hold decisions
    df = pd.DataFrame(index=[f"({qty}) {power}" for qty, power in deck_A], 
                    columns=[f"({qty}) {power}" for qty, power in deck_B], 
                    dtype='object')

    # Function to count outs
    def count_outs_bottom_mine(opp_power, my_deck):
        return sum(qty for qty, power in my_deck if opp_power < power)
    
    def count_outs_bottom_opp(my_power, opp_deck):
        return sum(qty for qty, power in opp_deck if my_power > power)
    
    # Populate the DataFrame with decisions
    for (my_qty, my_power) in deck_A:
        for (their_qty, their_power) in deck_B:
            if my_power > their_power:
                decision = 'win'  # No need to bottom any card, already winning
            else:
                outs_bottom_mine = count_outs_bottom_mine(their_power, deck_A)
                outs_bottom_opp = count_outs_bottom_opp(my_power, deck_B)

                # Decide based on which action gives more outs
                if outs_bottom_mine > outs_bottom_opp:
                    decision = 'mine'
                elif outs_bottom_mine < outs_bottom_opp:
                    decision = 'opp'
                else:
                    decision = 'either'


            df.loc[f"({my_qty}) {my_power}", f"({their_qty}) {their_power}"] = f"{decision}"

    return df


# Given decks
deck_A = [(14, 0), (2, 4), (3, 5), (15, 6), (6, 7), (11, 8), (3, 9), (3, 10)]
deck_B = [(8, 0), (16, 6), (33, 7), (3, 9)]

df = run_clash_sim(deck_A, deck_B)
df.to_csv('victor v kayo.csv')
df
```
